**Step 1:** Import the required libraries.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from google.colab import files
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

**Step 2:** Upload the dataset.

```python
uploaded = files.upload()
```

**Step 3:** Encode and train the data.

```python
dataframe = pd.read_csv('auto_mpg_dataset.csv')

X = dataframe[['cylinders','displacement','horsepower','weight','acceleration','model year','origin']]
y = dataframe['mpg']

label_encoder = LabelEncoder()
for column in X.columns:
    X[column] = label_encoder.fit_transform(X[column])

print (X)
print (y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)
```

**Step 4:** Finding MSE and R2 Score.

```python
# Initializing Linear regression model
lr = LinearRegression()

# Train the model on the training data
lr.fit(X_train, y_train)

# Predict on the test data
y_pred = lr.predict(X_test)

# Finding MSE and R2 Score
print("\nMean Squared Error (MSE):", metrics.mean_squared_error(y_test, y_pred))
print("R2 Score:", metrics.r2_score(y_test, y_pred))
```

**Step 5:** Plotting the graph and finding model accuracy.

```python
plt.plot(color='r', linestyle='--')
plt.scatter(y_test, y_pred)
plt.xlabel("Actual MPG")
plt.ylabel("Predicted MPG")
plt.title("Actual vs. Predicted MPG")
plt.show()

accuracy = lr.score(X_test, y_test)
accuracy_in_percentage = accuracy * 100
print("\nAccuracy:", accuracy)
print("\nAccuracy in %age:", accuracy_in_percentage)
```
