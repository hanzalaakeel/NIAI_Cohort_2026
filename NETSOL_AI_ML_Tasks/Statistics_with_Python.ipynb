## ⭐ Task 3 — Quartiles & Five-Number Summary

**Dataset:** Marks of 12 students:
```python
marks = [45, 52, 58, 61, 67, 72, 76, 80, 85, 91, 95, 99]
```

**Your job:**
1. Find **Q1**, **Q2 (median)**, **Q3**
2. Calculate the **IQR** (Q3 − Q1)
3. Print the full **five-number summary** (Min, Q1, Q2, Q3, Max)

**Expected output:**
```
=== Five-Number Summary ===
Min    : 45
Q1     : 60.25
Median : 74.0
Q3     : 85.75
Max    : 99
IQR    : 25.5
```

> 💡 **Hints:**
> - `np.percentile(data, 25)` → Q1
> - `np.percentile(data, 50)` → Q2
> - `np.percentile(data, 75)` → Q3
> - IQR = Q3 − Q1

````python
# Task 3 — Your code here
marks = [45, 52, 58, 61, 67, 72, 76, 80, 85, 91, 95, 99]

# Q1, Q2, Q3
q1 = np.percentile(marks, 25)
q2 = np.percentile(marks, 50)
q3 = np.percentile(marks, 75)

# IQR
iqr = q3-q1

# Five-number summary
print('=== Five-Number Summary ===')
print(f'Min    : {np.min(marks)}')
print(f'Q1     : {q1:.2f}')
print(f'Median : {q2:.1f}')
print(f'Q3     : {q3:.2f}')
print(f'Max    : {np.max(marks)}')
print(f'IQR    : {q3-q1}')
````

## ⭐ Task 4 — Detect Outliers Using IQR

**Dataset:** Annual salaries at a small company (one person is the CEO!):
```python
salaries = [30000, 32000, 35000, 33000, 200000, 31000, 34000, 36000, 29000, 32500]
```

**Your job:**
1. Compute Q1, Q3, and IQR
2. Compute the **outlier fences**:
   - Lower fence = Q1 − 1.5 × IQR
   - Upper fence = Q3 + 1.5 × IQR
3. Loop through salaries and **print any outliers**

**Expected output:**
```
Lower fence : 22,875.00
Upper fence : 43,625.00
🚨 Outlier detected: 200,000
```

> 💡 **Hints:**
> - Use `for s in salaries:` loop
> - Check: `if s < lower or s > upper:`
> - 200,000 (the CEO salary) should be the outlier
> - Use `f'{s:,}'` to format numbers with commas

````python
# Task 4 — Your code here
salaries = [30000, 32000, 35000, 33000, 200000,
            31000, 34000, 36000, 29000, 32500]

# Q1, Q3, IQR
q1 = np.percentile(salaries, 25)
q3 = np.percentile(salaries, 75)
iqr = q3-q1

# Outlier fences
lower = q1-1.5*iqr
upper = q3+1.5*iqr

# Detect and print outliers
print(f'Lower fence : {lower:,.2f}')
print(f'Upper fence : {upper:,.2f}')
````

## ⭐⭐ Task 5 — Z-Score Normalisation

**Dataset:** Heights (cm) of 10 people:
```python
heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]
```

**Your job:**
1. Compute the **mean** and **std** of heights
2. Calculate the **z-score** for each height
3. Print each height with its z-score
4. Flag anyone with **|z| > 1.5** as `'⚠️ Unusual'`

**Expected output:**
```
Height  Z-Score  Flag
 150     -1.84   ⚠️ Unusual
 162     -0.68
 170     +0.09
 ...
 180     +1.08
```

> 💡 **Hints:**
> - `z = (x - mean) / std`
> - Loop: `for h in heights:`
> - `flag = '⚠️ Unusual' if abs(z) > 1.5 else ''`
> - This is exactly what `StandardScaler` does in scikit-learn!

````python
# Task 5 — Your code here
heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]

# Compute mean and std
mean = np.mean(heights)
std = np.std(heights)

# Print table header
print(f'  {"Height":>6}  {"Z-Score":>8}  Flag')
print('-' * 30)

# Calculate and print z-scores
for h in heights:
    z = (h - mean) / std
    flag = '⚠️ Unusual' if abs(z) > 1.5 else ''
    print(f'{h:>8}  {z:>+8.2f}  {flag}')
````

## ⭐⭐ Task 6 — Explore Skewness

**Create these two datasets:**
```python
right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]
```

**Your job:**
1. Compute **skewness** for both datasets using `stats.skew()`
2. Compute **mean** and **median** for both
3. Print results and verify the rule:
   - Right skew → mean **>** median
   - Left skew  → mean **<** median

**Expected output:**
```
--- Right Skewed ---
Skewness: +1.56   direction: positive →
Mean: 7.73   Median: 4.0
Mean > Median ✓  (right skew rule holds!)

--- Left Skewed ---
Skewness: -1.38   direction: negative ←
Mean: 13.0   Median: 16.0
Mean < Median ✓  (left skew rule holds!)
```

> 💡 **Hints:**
> - `stats.skew(data)` returns a number
> - Positive = right skew, Negative = left skew
> - Check the mean vs median relationship each time

````python
# Task 6 — Your code here
right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]

# For right_skewed:
skew_r = stats.skew(right_skewed)
mean_r = np.mean(right_skewed)
median_r = np.median(right_skewed)

# For left_skewed:
skew_l = stats.skew(left_skewed)
mean_l = np.mean(left_skewed)
median_l = np.median(left_skewed)

# Print and verify the skew rule
print('--- Right Skewed ---')
print(f'Skewness: {skew_r:+.2f}   direction: positive →')
print(f'Mean: {mean_r:.2f}   Median: {median_r:.1f}')
if mean_r > median_r:
    print('Mean > Median ✓  (right skew rule holds!)')
else:
    print('Mean <= Median ✗  (right skew rule fails!)')

print('\n--- Left Skewed ---')
print(f'Skewness: {skew_l:+.2f}   direction: negative ←')
print(f'Mean: {mean_l:.1f}   Median: {median_l:.1f}')
if mean_l < median_l:
    print('Mean < Median ✓  (left skew rule holds!)')
else:
    print('Mean >= Median ✗  (left skew rule fails!)')
````

## ⭐⭐ Task 7 — Correlation Analysis

**Dataset:** Hours of sleep and test scores for 10 students:
```python
hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]
```

**Your job:**
1. Calculate the **Pearson correlation** and **p-value**
2. Determine strength: Strong (>0.8) / Moderate (>0.5) / Weak
3. State if the result is statistically significant (p < 0.05)
4. **Bonus:** Plot a scatter plot with a trend line

**Expected output:**
```
Pearson r = 0.997
p-value   = 0.000001
Strength : STRONG positive correlation
Significant: YES ✅
```

> 💡 **Hints:**
> - `corr, pval = stats.pearsonr(x, y)`
> - Use `if/elif/else` for strength labels
> - Bonus: `plt.scatter(x, y)` then `np.polyfit(x, y, 1)` for trend line

````python
# Task 7 — Your code here
hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]

# 1. Pearson correlation
corr, pval = stats.pearsonr(hours_sleep, test_score)
print(f'Pearson r = {corr:.3f}')
print(f'p-value   = {pval:.6f}')

# 2. Strength classification
strength = ''
if corr > 0.8:
    strength = 'STRONG positive correlation'
elif corr > 0.5:
    strength = 'MODERATE positive correlation'
elif corr < -0.8:
    strength = 'STRONG negative correlation'
elif corr < -0.5:
    strength = 'MODERATE negative correlation'
elif abs(corr) <= 0.5:
    strength = 'WEAK correlation'
print(f'Strength : {strength}')

# 3. Significance check
significant = 'YES ✅' if pval < 0.05 else 'NO ❌'
print(f'Significant: {significant}')

# 4. (Bonus) Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(hours_sleep, test_score, label='Data Points')

# Add trend line
m, c = np.polyfit(hours_sleep, test_score, 1) # m=slope, c=intercept
plt.plot(hours_sleep, [m*h + c for h in hours_sleep], color='red', linestyle='--', label=f'Trend Line (y={m:.2f}x + {c:.2f})')

plt.title('Hours of Sleep vs. Test Score')
plt.xlabel('Hours of Sleep')
plt.ylabel('Test Score')
plt.grid(True)
plt.legend()
plt.show()
````

## ⭐⭐⭐ Task 9 — Hypothesis Testing

**Scenario:** A school tested a new teaching method.

```python
old_method = [65, 70, 68, 72, 74, 69, 71, 73, 66, 75]
new_method = [78, 82, 80, 85, 79, 83, 81, 84, 77, 86]
```

**Your job:**
1. State **H₀** and **H₁** in plain English (as print statements)
2. Run an **independent samples t-test**
3. Print t-statistic and p-value
4. Print the means of both groups
5. Write a clear **conclusion** based on the p-value

**Expected output:**
```
H₀: Both methods produce the same results
H₁: The new method is significantly different

Old method mean: 70.3
New method mean: 81.5
t-statistic    : -10.470
p-value        : 0.000000

✅ Reject H₀ — new method is significantly better!
```

> 💡 **Hints:**
> - `stats.ttest_ind(group1, group2)`
> - If p < 0.05 → reject H₀
> - Compare the means to see which group performed better
> - The t-stat being negative just means group1 < group2

````python
# Task 9 — Your code here
old_method = [65, 70, 68, 72, 74, 69, 71, 73, 66, 75]
new_method = [78, 82, 80, 85, 79, 83, 81, 84, 77, 86]

# 1. State hypotheses
print('H₀: Both methods produce the same results')
print('H₁: The new method is significantly different')
print('\n')

# 2. Run t-test
t_statistic, p_value = stats.ttest_ind(old_method, new_method)

# 3. Print t-stat and p-value
print(f'Old method mean: {np.mean(old_method):.1f}')
print(f'New method mean: {np.mean(new_method):.1f}')
print(f't-statistic    : {t_statistic:.3f}')
print(f'p-value        : {p_value:.6f}')
print('\n')

# 5. Conclusion
if p_value < 0.05:
    if np.mean(new_method) > np.mean(old_method):
        print('✅ Reject H₀ — new method is significantly better!')
    else:
        print('✅ Reject H₀ — new method is significantly worse!')
else:
    print('❌ Fail to reject H₀ — no significant difference found.')
````

## ⭐⭐⭐ Task 10 — MINI PROJECT: Build a Stats Report Function 🏆

**Write a function `stats_report(data, name)` that prints a complete statistical summary.**

It should output:
1. Dataset name and values
2. Mean, Median, Mode
3. Std Dev and Variance
4. Q1, Q2, Q3 and IQR
5. Skewness with direction label
6. Any outliers (using IQR method)
7. Z-scores — flag any value where |z| > 2

**Test it with:**
```python
stats_report([12, 15, 14, 10, 100, 13, 14, 15, 16, 11], 'Sample Data')
```

**Expected output:**
```
========================================
STATS REPORT: Sample Data
========================================
Values    : [12, 15, 14, 10, 100, 13, 14, 15, 16, 11]

--- Central Tendency ---
Mean      : 22.0
Median    : 14.0
Mode      : 14 (appeared 2x)

--- Spread ---
Std Dev   : 26.41
Variance  : 697.8

--- Quartiles ---
Q1=11.75  Q2=14.0  Q3=15.25  IQR=3.5

--- Shape ---
Skewness  : +3.08  (right skewed)

--- Outliers (IQR method) ---
🚨 Outlier: 100

--- Z-Scores ---
  12   z=+0.38
  15   z=+0.49
 ...
 100   z=+2.95  ⚠️ EXTREME
========================================
```

> 💡 **Hints:**
> - Use `def stats_report(data, name):`
> - Call all the numpy/scipy functions inside the function
> - Use `stats.mode(data).mode` and `.count` for mode
> - Build it step by step — get one section working, then add the next

> 🚀 **Bonus Challenges:**
> - Make the function **return a dictionary** instead of just printing
> - Add a **histogram** chart of the data
> - Accept a **pandas Series** as input
> - Test it on the `scores` and `salaries` datasets from earlier tasks

> 🎯 If you can build this function from scratch, you truly understand statistics for ML!

````python
# Task 10 — Your code here
# Build your stats_report function below

def stats_report(data, name):
    print('=' * 40)
    print(f'STATS REPORT: {name}')
    print('=' * 40)

    # 1. Dataset name and values
    print(f'Values    : {data}')

    # 2. Central Tendency (mean, median, mode)
    mean_val = np.mean(data)
    median_val = np.median(data)
    # Ensure mode_result.mode is always an array by setting keepdims=True
    mode_result = stats.mode(data, keepdims=True)

    # stats.mode returns ModeResult object where .mode and .count are always arrays
    if len(mode_result.mode) > 0:
        mode_val = mode_result.mode[0] # Assuming we want the first mode if multiple
        mode_count = mode_result.count[0]
    else:
        # This branch should theoretically not be reached if data is not empty,
        # as stats.mode with keepdims=True will always return an array.
        mode_val = 'N/A'
        mode_count = 0

    print('\n--- Central Tendency ---')
    print(f'Mean      : {mean_val:.2f}')
    print(f'Median    : {median_val:.1f}')
    if mode_val != 'N/A':
        print(f'Mode      : {mode_val} (appeared {mode_count}x)')
    else:
        print(f'Mode      : {mode_val}')

    # 3. Spread (std, variance)
    std_dev = np.std(data)
    variance_val = np.var(data)
    print('\n--- Spread ---')
    print(f'Std Dev   : {std_dev:.2f}')
    print(f'Variance  : {variance_val:.2f}')

    # 4. Quartiles (Q1, Q2, Q3, IQR)
    q1 = np.percentile(data, 25)
    q2 = np.percentile(data, 50) # Median
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    print('\n--- Quartiles ---')
    print(f'Q1={q1:.2f}  Q2={q2:.1f}  Q3={q3:.2f}  IQR={iqr:.2f}')

    # 5. Skewness with direction label
    skewness = stats.skew(data)
    skew_direction = ''
    if skewness > 0.5: # Using 0.5 as a threshold for noticeable skew
        skew_direction = '(right skewed)'
    elif skewness < -0.5:
        skew_direction = '(left skewed)'
    else:
        skew_direction = '(symmetrical)'
    print('\n--- Shape ---')
    print(f'Skewness  : {skewness:+.2f}  {skew_direction}')

    # 6. Outliers (IQR method)
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    outliers = [x for x in data if x < lower_fence or x > upper_fence]
    print('\n--- Outliers (IQR method) ---')
    if outliers:
        for outlier in outliers:
            print(f'🚨 Outlier: {outlier}')
    else:
        print('No outliers detected.')

    # 7. Z-scores — flag any value where |z| > 2
    mean_for_z = np.mean(data)
    std_for_z = np.std(data)
    print('\n--- Z-Scores ---')
    if std_for_z == 0: # Avoid division by zero if all data points are identical
        print('Cannot compute Z-scores: Standard deviation is zero.')
    else:
        for x in data:
            z = (x - mean_for_z) / std_for_z
            flag = '⚠️ EXTREME' if abs(z) > 2 else ''
            print(f'  {x:>3}   z={z:+.2f}  {flag}')

    print('=' * 40)


# Test it!
stats_report([12, 15, 14, 10, 100, 13, 14, 15, 16, 11], 'Sample Data')
````
